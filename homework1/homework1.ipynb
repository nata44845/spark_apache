```python
import pyspark
from pyspark import SparkContext
```

```python
# разбиение данных на упорядоченные последовательности
data = [45, 14, 1, 200, 58, 112, 34, 225, 300, 500, 600, 17, -5, 32, 56, 67, 78]
sequences = []
seq = [data[0]]
for x in data[1:]:
    if x >= seq[-1]:
        seq.append(x)
    else:
        sequences.append(seq)
        seq = [x]
if seq:
    sequences.append(seq)

print(sequences)

# последовательности максимальной длины
sc = SparkContext("local", "LongestSequence")
rdd = sc.parallelize(sequences)
num = len(rdd.max(key=lambda x: len(x)))
print(f"{num=}")
longest = rdd.filter(lambda x: len(x)>=num).collect()
print(longest)
```

```python
sc.stop()
```
