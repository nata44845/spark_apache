```python
data = {'id': [1, 2, 3, 4], 'Name': ['Joe', 'Henry', 'Sam', 'Max'], 'Salary': ['70', '80', '60', '90'], 'ManagerId': ['3', '4', 'Null', 'Null']}
```

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
import pandas as pd
```

```python
spark = SparkSession.builder.appName("Homework2").getOrCreate()
df = spark.createDataFrame(pd.DataFrame(data))
df.show()
```

```python
joined_df = df.alias("employees").join(df.alias("managers"), F.col("employees.ManagerId") == F.col("managers.id"), "left")
joined_df.show()
```

```python
result_df = joined_df.filter(F.col("employees.Salary") > F.col("managers.Salary"))
result_df.select("employees.Name").show()
```
